```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = list(stopwords.words('english'))
```

removing stopwords

```python
new_text=[]
for line in data.Text:
  filtered=[]
  small=line.lower()
  list_of_words=small.split()
  for word in list_of_words:
    if word not in stop_words:
      filtered.append(word)
  filtered_string=""
  for i in filtered:
      filtered_string+=i+" "
  new_text.append(filtered_string)
```

regular expressions

```python
import re
new_text2=[]
for line in new_text:
  text=re.sub(r"[^A-Za-z]"," ",line)
  new_text2.append(text)
```

```python
new_text2[:5]
```

stemming (general form of words)

```python
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lem=WordNetLemmatizer()
```

```python
new_text3=[]
for line in new_text2:
  list1=[]
  s=line.split()
  for j in s:
    a=lem.lemmatize(j)
    list1.append(a)
  new_text3.append(" ".join(list1))
```

```python
new_text3[:5]
```

```python
data.Text=new_text3
```

```python
x=data.Text
y=data.Label
```

vectorizing sentences

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec=CountVectorizer()
```

```python
vec_x=vec.fit_transform(x)
```

logistic regression

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
log.fit(vec_x,y)
```

```python
from sklearn.metrics import f1_score
```

```python
pred_y=log.predict(vec_x)
```
